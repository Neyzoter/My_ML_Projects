```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 球员基础数据
advanced_basic = pd.read_csv("./NBAdata/advanced_basic.csv", sep=',')
```

```python
advanced_basic.head(10)
```

```python
advanced_basic_harden = advanced_basic[advanced_basic.iloc[:,0]=="James Harden"]
advanced_basic_harden
```

```python
X_Label = advanced_basic_harden.columns[1]
Y_Label = advanced_basic_harden.columns[16]
x = range(len(advanced_basic_harden.iloc[:,1]))
# 需要反序[::-1]
plt.plot(x,(advanced_basic_harden.iloc[:,16])[::-1])
plt.xticks(x, (advanced_basic_harden.iloc[:,1])[::-1], rotation=45)
plt.margins(0.08)
plt.subplots_adjust(bottom=0.15)
plt.title("Harden's "+Y_Label)
plt.xlabel(X_Label)
plt.ylabel(Y_Label)
plt.show()
```
